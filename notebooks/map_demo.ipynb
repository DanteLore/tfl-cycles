```python
import geopandas as gpd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from shapely.geometry import Point
import mplleaflet

# set the filepath and load in a shapefile
SHAPE_FILE = "../data/maps/statistical-gis-boundaries-london/ESRI/London_Borough_Excluding_MHW.shp"
BIKE_POINTS_FILE = "../data/bike-points.csv"

crs = {'init': 'EPSG:27700'}

map_df = gpd.read_file(SHAPE_FILE)

fig, ax = plt.subplots(1, figsize=(18, 14))
ax.axis('off')
map_df.plot(linewidth=0.8, ax=ax, edgecolor='0.8')

bp_df = pd.read_csv(BIKE_POINTS_FILE)
geometry = [Point(xy) for xy in zip(bp_df['osgb_x'], bp_df['osgb_y'])]

gdf = gpd.GeoDataFrame(bp_df, crs=crs, geometry=geometry)
gdf.plot(ax=ax, markersize=10, color='Red', marker='o')
```
